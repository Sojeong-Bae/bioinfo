````
import collections
 
from scipy.stats import entropy
  
def estimate_shannon_entropy(dna_sequence):
    bases = collections.Counter([tmp_base for tmp_base in dna_sequence])
    # define distribution
    dist = [x/sum(bases.values()) for x in bases.values()]
 
    # use scipy to calculate entropy
    entropy_value = entropy(dist, base=2)
 
    return entropy_value
````

````
pileup_long_read_data['entropy'] = pileup_long_read_data['matches'].map(lambda x : estimate_shannon_entropy(x))
pileup_long_read_data
````

````
final_chr1_bedgraph_pileup_data = chr1_bedgraph_pileup_data[(chr1_bedgraph_pileup_data['entropy'] >= 1.0)]
final_chr1_bedgraph_pileup_data.head()
````

````
import pandas as pd

pileup = pd.read_csv('CLIP-35L33G-chr1.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup.tail()
````

````
import re
toremove = re.compile(r'[<>$*#]|\^.')
pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))
````

````
pileup[['chrom', 'pos', 'matches']]
````

````
from scipy.stats import entropy
import seaborn as sns

def shannon_entropy(matches: str):
  counts = {"A" : 0, "T" : 0, "G" : 0, "C" : 0}
  for base in matches:
    try :
      counts[base.upper()] += 1
    except KeyError:
      continue
  return 0 if sum(counts.values()) <= 0 else entropy([x/sum(counts.values()) for x in counts.values()])

bedgraph = pileup[["chrom", "pos"]].copy()
bedgraph["start"] = bedgraph["pos"]
bedgraph["end"] = bedgraph["start"] + 1
bedgraph["value"] = pileup["matches"].apply(shannon_entropy)
bedgraph = bedgraph[["chrom", "start", "end",  "value"]]
bedgraph
````
